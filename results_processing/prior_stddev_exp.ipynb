```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Sample data for demonstration
prior_std_dev = np.array([0.001, 0.01, 0.05, 0.1, 0.2, 0.3, 0.5, 1.0, 2.0, 5.0, 7.5, 10.0])
x_indices = np.arange(len(prior_std_dev))
nll = np.ones(np.size(prior_std_dev))  # Random data for NLL
accuracy = np.ones(np.size(prior_std_dev))  # Random data for Accuracy (0 to 1 scale)
ece = np.ones(np.size(prior_std_dev))  # Random data for ECE

```

```python
colors = [
    '#1f77b4',  # Blue
    '#ff7f0e',  # Orange
    '#2ca02c',  # Green
    '#d62728',  # Red
    '#9467bd',  # Purple
    '#8c564b',  # Brown
    '#e377c2',  # Pink
    '#7f7f7f',  # Gray
    '#bcbd22',  # Olive
    '#17becf',  # Teal
    '#ffbb78',  # Light Orange
    '#98df8a'   # Light Green
]
```

```python
plt.style.use('fivethirtyeight')
# colors = plt.cm.rainbow(np.linspace(0, 1, len(prior_std_dev)))
# colors = np.random.rand(len(prior_std_dev), 3)
# # Scatter plots
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# NLL Scatter plot
axs[0].scatter(x_indices, nll, color=colors, s=100)
#axs[0].set_title('Prior Standard Deviation vs NLL')
axs[0].set_xlabel('Prior Standard Deviation', fontsize=14)
axs[0].set_xticks(x_indices)
axs[0].set_xticklabels(prior_std_dev, rotation=45)
axs[0].set_ylabel('NLL', fontsize=14)

# Accuracy Scatter plot
axs[1].scatter(x_indices, accuracy, color=colors, s=100)
# axs[1].set_title('Prior Standard Deviation vs Accuracy')
axs[1].set_xlabel('Prior Standard Deviation', fontsize=14)
axs[1].set_xticks(x_indices)
axs[1].set_xticklabels(prior_std_dev, rotation=45)
axs[1].set_ylabel('Accuracy', fontsize=14)

# ECE Scatter plot
axs[2].scatter(x_indices, ece, color=colors, s=100)
# axs[2].set_title('Prior Standard Deviation vs ECE')
axs[2].set_xlabel('Prior Standard Deviation', fontsize=14)
axs[2].set_xticks(x_indices)
axs[2].set_xticklabels(prior_std_dev, rotation=45)
axs[2].set_ylabel('ECE', fontsize=14)

plt.tight_layout()
plt.show()
```
