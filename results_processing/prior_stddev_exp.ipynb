```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Sample data for demonstration
prior_std_dev = np.array([0.001, 0.01, 0.05, 0.1, 0.2, 0.3, 0.5, 1.0, 2.0, 5.0, 7.5, 10.0])
x_indices = np.arange(len(prior_std_dev))
nll = np.array([0.1687, 0.1733, 0.1552, 0.1572, 0.1624, 0.1685, 0.1703, 0.1801, 0.1769, 0.1764, 0.1748, 0.1829])   
accuracy = np.array([96.03, 95.96, 95.9, 96.02, 96.05, 96.03, 95.75, 95.76, 95.79, 96.17, 95.83, 95.79])  
ece = np.array([0.02458, 0.02532, 0.02339, 0.02211, 0.02197, 0.02174, 0.02553, 0.02555, 0.02419, 0.02061, 0.02494, 0.02499])  

nll_v2 = np.array([0.1742, 0.1723, 0.1650, 0.1671, 0.1639, 0.1664, 0.1798, 0.1725, 0.1817, 0.1794, 0.1792, 0.1805])
accuracy_v2 = np.array([95.72, 95.72, 96.1, 95.92, 95.79, 95.91, 95.61, 96.04, 95.96, 95.79, 95.84, 95.63])
ece_v2 = np.array([0.02716, 0.02622, 0.02302, 0.02364, 0.02383, 0.02327, 0.02735, 0.02291, 0.02371, 0.02567, 0.02463, 0.02616])

```

```python
colors = [
    '#1f77b4',  # Blue
    '#ff7f0e',  # Orange
    '#2ca02c',  # Green
    '#d62728',  # Red
    '#9467bd',  # Purple
    '#8c564b',  # Brown
    '#e377c2',  # Pink
    '#7f7f7f',  # Gray
    '#bcbd22',  # Olive
    '#17becf',  # Teal
    '#ffbb78',  # Light Orange
    '#98df8a'   # Light Green
]
```

```python
plt.style.use('default')
colors = plt.cm.gist_rainbow(np.linspace(0, 1, len(prior_std_dev)))
# colors = np.random.rand(len(prior_std_dev), 3)
# # Scatter plots
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# NLL Scatter plot
axs[0].scatter(x_indices, nll_v2, color=colors, s=100)
#axs[0].set_title('Prior Standard Deviation vs NLL')
axs[0].set_xlabel('Prior Standard Deviation', fontsize=14)
axs[0].set_xticks(x_indices)
axs[0].set_xticklabels(prior_std_dev, rotation=45)
axs[0].set_ylabel('NLL', fontsize=16)
axs[0].grid(alpha=0.6)
axs[0].tick_params(axis='y', labelsize=14)
axs[0].tick_params(axis='x', labelsize=11)

# Accuracy Scatter plot
axs[1].scatter(x_indices, accuracy_v2, color=colors, s=100)
# axs[1].set_title('Prior Standard Deviation vs Accuracy')
axs[1].set_xlabel('Prior Standard Deviation', fontsize=14)
axs[1].set_xticks(x_indices)
axs[1].set_xticklabels(prior_std_dev, rotation=45)
axs[1].set_ylabel('Accuracy', fontsize=16)
axs[1].grid(alpha=0.6)
axs[1].tick_params(axis='y', labelsize=14)
axs[1].tick_params(axis='x', labelsize=11)

# ECE Scatter plot
axs[2].scatter(x_indices, ece_v2, color=colors, s=100)
# axs[2].set_title('Prior Standard Deviation vs ECE')
axs[2].set_xlabel('Prior Standard Deviation', fontsize=14)
axs[2].set_xticks(x_indices)
axs[2].set_xticklabels(prior_std_dev, rotation=45)
axs[2].set_ylabel('ECE', fontsize=16)
axs[2].grid(alpha=0.6)
axs[2].tick_params(axis='y', labelsize=14)
axs[2].tick_params(axis='x', labelsize=11)

plt.tight_layout()
plt.show()
```
