```python
import matplotlib.pyplot as plt
import numpy as np

# Read entropy values from the text file
file_path = 'BNN_normal_data_entropy_v2.txt'

with open(file_path, 'r') as file:
    entropies = [float(line.strip()) for line in file]
entropies = np.log(entropies)

# Create a histogram
plt.figure(figsize=(10, 6))
plt.hist(entropies, bins=40, edgecolor='black')
plt.title('Histogram of Entropy Values')
plt.xlabel('Entropy')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Read data from text files
file_path_bnn = 'BNN_corrupt_data_entropy_v2.txt'
file_path_det = 'det_corrupt_data_entropy.txt'
bnn = np.loadtxt(file_path_bnn)
det = np.loadtxt(file_path_det)

# Plot histograms
plt.figure(figsize=(10, 6))
plt.hist(bnn, bins=200, alpha=0.6, label='BNN entropies', color='b', edgecolor='black')
plt.hist(det, bins=200, alpha=0.4, label='Deterministic entropies', color='r', edgecolor='black')



# Add title and labels
plt.title('Histograms of Data 1 and Data 2')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Add legend
plt.legend()

# Show plot
plt.show()
```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('ggplot')
file_path_bnn1 = 'BNN_corrupt_data_entropy_log2.txt'
# file_path_bnn2 = 'BNN_corrupt_data_entropy_NewInit_log2.txt'
file_path_bnn2 = 'BNN_corrupt_data_entropy_v3.txt'
file_path_det = 'det_corrupt_data_entropy_log2.txt'
file_path_ens = 'ens_corrupt_data_entropy.txt'
bnn1 = np.loadtxt(file_path_bnn1)
bnn2 = np.loadtxt(file_path_bnn2)
det = np.loadtxt(file_path_det)
ens = np.loadtxt(file_path_ens)

min_val = min(np.min(bnn1), np.min(bnn2), np.min(det), np.min(ens))
max_val = max(np.max(bnn1), np.max(bnn2), np.max(det) , np.max(ens))
bins = np.linspace(np.log10(min_val), np.log10(max_val), 70)

bw_adjust = 0.5
sns.histplot(bnn1, bins=bins, log_scale=(True,False), alpha=0.4, color='#1f77b4', kde=False, kde_kws={'bw_adjust': bw_adjust})
sns.histplot(bnn2, bins=bins, log_scale=(True,False), alpha=0.4, color='#ff7f0e', kde=False, kde_kws={'bw_adjust': bw_adjust})
sns.histplot(det, bins=bins, log_scale=(True,False), alpha=0.4, color='#2ca02c', kde=False, kde_kws={'bw_adjust': bw_adjust})
sns.histplot(ens, bins=bins, log_scale=(True,False), alpha=0.4, color='red', kde=False, kde_kws={'bw_adjust': bw_adjust})
plt.gca().set(title='Corrupted Data Entropy Distribution', xlabel='Entropy', ylabel='Frequency')
plt.legend(['Rank1 BNN1', 'Rank1 BNN2', 'WRN-28-10', 'Ensemble'])
plt.show()

sns.kdeplot(bnn1, log_scale=(True,False), color='#1f77b4', bw_adjust=bw_adjust)
sns.kdeplot(bnn2, log_scale=(True,False), color='#ff7f0e', bw_adjust=bw_adjust)
sns.kdeplot(det, log_scale=(True,False), color='#2ca02c', bw_adjust=bw_adjust)
sns.kdeplot(ens, log_scale=(True,False), color='red', bw_adjust=bw_adjust)
plt.gca().set(title='Corrupted Data Entropy Distribution', xlabel='Entropy', ylabel='Density')
plt.legend(['Rank1 BNN1', 'Rank1 BNN2', 'WRN-28-10', 'Ensemble'])
plt.show()
```

```python

# Set font to Computer Modern 10pt


plt.style.use('ggplot')

file_path_bnn1 = 'BNN_corrupt_data_entropy_log2.txt'
file_path_bnn2 = 'BNN_corrupt_data_entropy_v3.txt'
file_path_det = 'det_corrupt_data_entropy_log2.txt'
file_path_ens = 'ens_corrupt_data_entropy.txt'

bnn1 = np.loadtxt(file_path_bnn1)
bnn2 = np.loadtxt(file_path_bnn2)
det = np.loadtxt(file_path_det)
ens = np.loadtxt(file_path_ens)

min_val = min(np.min(bnn1), np.min(bnn2), np.min(det), np.min(ens))
max_val = max(np.max(bnn1), np.max(bnn2), np.max(det), np.max(ens))
bins = np.linspace(np.log10(min_val), np.log10(max_val), 70)

bw_adjust = 0.5

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Histogram subplot
sns.histplot(bnn1, bins=bins, log_scale=(True,False), alpha=0.4, color='#1f77b4', kde=False, kde_kws={'bw_adjust': bw_adjust}, ax=axes[0])
sns.histplot(bnn2, bins=bins, log_scale=(True,False), alpha=0.4, color='#ff7f0e', kde=False, kde_kws={'bw_adjust': bw_adjust}, ax=axes[0])
sns.histplot(det, bins=bins, log_scale=(True,False), alpha=0.4, color='#2ca02c', kde=False, kde_kws={'bw_adjust': bw_adjust}, ax=axes[0])
sns.histplot(ens, bins=bins, log_scale=(True,False), alpha=0.4, color='red', kde=False, kde_kws={'bw_adjust': bw_adjust}, ax=axes[0])
axes[0].set(title='Histogram of Prediction Entropy Distribution', xlabel='Entropy', ylabel='Frequency')
axes[0].set_title('Histogram of Prediction Entropy Distribution (OOD)', fontsize=15, fontname='cmr10')
axes[0].legend(['Rank1 BNN V1', 'Rank1 BNN V2', 'Det. WRN-28-10', 'Deep Ensemble'])

# KDE subplot
sns.kdeplot(bnn1, log_scale=(True,False), color='#1f77b4', bw_adjust=bw_adjust, ax=axes[1], alpha=0.6, linewidth=2.5)
sns.kdeplot(bnn2, log_scale=(True,False), color='#ff7f0e', bw_adjust=bw_adjust, ax=axes[1], alpha=0.6, linewidth=2.5)
sns.kdeplot(det, log_scale=(True,False), color='#2ca02c', bw_adjust=bw_adjust, ax=axes[1], alpha=0.6, linewidth=2.5)
sns.kdeplot(ens, log_scale=(True,False), color='red', bw_adjust=bw_adjust, ax=axes[1], alpha=0.6, linewidth=2.5)
axes[1].set(title='KDE of Prediction Entropy Distribution', xlabel='Entropy', ylabel='Density')
axes[1].set_title('KDE of Prediction Entropy Distribution (OOD)', fontsize=15, fontname='cmr10')
axes[1].legend(['Rank1 BNN V1', 'Rank1 BNN V2', 'Det. WRN-28-10', 'Deep Ensemble'])

plt.tight_layout()
plt.show()
```
