```python
import matplotlib.pyplot as plt

# Read entropy values from the text file
file_path = 'BNN_normal_data_entropy.txt'

with open(file_path, 'r') as file:
    entropies = [float(line.strip()) for line in file]

# Create a histogram
plt.figure(figsize=(10, 6))
plt.hist(entropies, bins=80, edgecolor='black')
plt.title('Histogram of Entropy Values')
plt.xlabel('Entropy')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
